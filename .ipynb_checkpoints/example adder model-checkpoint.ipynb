```python
import numpy as np
```

```python
class ApproximateSummator:
    def __init__(self): 
        self.dataLabels=[(i,j,i+j) for i in range(10) for j in range(10)] ##for next line  ##個位數相加
        self.dataLabels=[(self.settle(i),self.settle(j),k) for i,j,k in self.dataLabels] #true dataLables (one-hot encoding)(最後一個int是答案)
        self.theta1=np.random.uniform(0,1,10) #產生0~1之間的小數 十個為一個array 
        self.theta2=np.random.uniform(0,1,10)
        self.bias=0 
    def settle(self,x): #ont-hot encoding的概念
        data=np.zeros(10) 
        data[x]=1 
        return data
    def train(self):
        for i in range(100):
            #if i%100==0: print (i)
            np.random.shuffle(self.dataLabels)
            for x1,x2,y in self.dataLabels: 
                out=np.dot(self.theta1,x1)+np.dot(self.theta2,x2)+self.bias
                err=(y-out)
                if i%100==0:
                    print(i,err)
                err*=0.1 ####這句開始不知道放哪
                self.theta1+=err*x1
                self.theta2+=err*x2 
                self.bias+=err
    def add(self,x1,x2):
        r=np.dot(self.theta1,self.settle(x1))+np.dot(self.theta2,self.settle(x2))+self.bias
        return r
    
```

```python
if __name__=="__main__": 
        ass=ApproximateSummator()
        ass.train()
right=0.0
for i in range(10): 
    for j in range(10): 
        result=ass.add(i, j)
        if result-(i+j)<10e-10: 
            right+=1
            print (i,"+",j,"=",result)
        else:
            #not right
            print (i,"+",j,"=",result,"NR")
print ("accuracy:",right,"%")
```
